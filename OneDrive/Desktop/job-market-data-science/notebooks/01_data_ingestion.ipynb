```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("../data/raw_jobs.csv")
df.head()
```

```python
df.shape
df.columns
df.info()
```

```python
df.columns = (
    df.columns
    .str.lower()
    .str.strip()
    .str.replace(" ", "_")
)
```

```python
df.columns.tolist()
```

```python
# dropping the junk column
df = df.drop(columns=['unnamed:_0'])
```

```python
# we have two 'description' columns 
df[['description', 'descriptions']].head()
```

```python
df = df.rename(columns={
    'title': 'job_title',
    'descriptions': 'job_description'
})
```

```python
# keeping the relevant columns 
important_cols = [
    'job_title',
    'company',
    'location',
    'salary',
    'job_description'
]

df = df[important_cols]
```

```python
# verification 
df.columns
df.head()
```

```python
# saving stage 1 clean data 
df.to_csv("data/clean_jobs_stage1.csv", index=False)
```
