```python
import pandas as pd
import re
from collections import Counter
```

```python
# Load clean data
df = pd.read_csv("../data/clean_jobs_stage1.csv")
```

```python
# Define key skills
skills = ['python', 'sql', 'excel', 'aws', 'machine learning', 'power bi', 'statistics']
```

```python
# Function to extract skills from a text
def extract_skills(text):
    text = str(text).lower()
    found_skills = [skill for skill in skills if re.search(r'\b' + re.escape(skill) + r'\b', text)]
    return found_skills
```

```python

# Apply function to job descriptions
df['skills_found'] = df['job_description'].apply(extract_skills)
df['skill_count'] = df['skills_found'].apply(len)
```

```python
# Top skills frequency
all_skills = [skill for sublist in df['skills_found'] for skill in sublist]
skill_freq = Counter(all_skills)
print(skill_freq.most_common(10))
```

```python

# Save updated dataset for salary analysis
df.to_csv("../data/clean_jobs_skills.csv", index=False)
```
